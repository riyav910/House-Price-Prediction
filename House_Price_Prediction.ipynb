#importing dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
from sklearn.metrics import accuracy_score
```

#Splitting the data and the target

```python
X = house_price_dataframe.drop(['Price'], axis = 1)
Y = house_price_dataframe['Price']
```

```python
print(X)
print(Y)
```

Splitting the data into Training Data ans Testing Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.15, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

#Model Training

XGBoost Regression

```python
#loading the model
model = XGBRegressor()
```

```python
from sklearn.base import BaseEstimator
```

```python
import xgboost as xgb
xg_reg = xgb.XGBRegressor(objective ='reg:linear', colsample_bytree = 1,eta=0.3, learning_rate = 0.1, max_depth = 5, alpha = 10, n_estimators = 2000)
```

```python
#training the model with training data
model.fit(X_train, Y_train)
```

#Evaluation

prediction on training data

```python
#accuracy for prediction on training data
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

prediction on testing data

```python
#accuracy for prediction on testing data
testing_data_prediction = model.predict(X_test)
```

```python
#graphical evaluation
grp = pd.DataFrame({'prediction':testing_data_prediction,'Actual':Y_test})
grp = grp.reset_index()
grp = grp.drop(['index'],axis=1)
plt.figure(figsize=(15,7.5))
plt.plot(grp[:120],linewidth=2)
plt.legend(['Actual','Predicted'],prop={'size': 20})
```
